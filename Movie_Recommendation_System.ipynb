## Project Overview

This project implements a content-based movie recommendation system using artificial intelligence and machine learning techniques. The system analyzes movie genres and descriptions using Term Frequency–Inverse Document Frequency (TF-IDF). It then calculates cosine similarity between movies to generate personalized recommendations based on a movie selected by the user.

The system includes an interactive interface that allows the user to select a movie, choose the desired number of recommendations, and generate a ranked recommendation list.

```python
# Replace missing values with empty strings
movies["genre"] = movies["genre"].fillna("")
movies["description"] = movies["description"].fillna("")

# Combine the movie genre and description
movies["combined_features"] = (
    movies["genre"] + " " + movies["description"]
)

movies[["title", "combined_features"]].head()
```

## 3. Building the Recommendation Model

The system uses TF-IDF vectorization and cosine similarity.

TF-IDF assigns numerical importance to words based on how frequently they appear in one movie's features compared with their frequency across the entire dataset. Cosine similarity then measures the similarity between the resulting movie vectors. A higher cosine-similarity score indicates that two movies have more content characteristics in common.

```python
# Create the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(
    stop_words="english",
    lowercase=True
)

# Convert the combined movie features into numerical vectors
tfidf_matrix = tfidf_vectorizer.fit_transform(
    movies["combined_features"]
)

print("TF-IDF matrix created successfully.")
print("TF-IDF matrix dimensions:", tfidf_matrix.shape)
```

```python
# Calculate similarity between every pair of movies
cosine_similarity_matrix = cosine_similarity(tfidf_matrix)

print("Cosine similarity matrix created successfully.")
print(
    "Cosine similarity matrix dimensions:",
    cosine_similarity_matrix.shape
)
```

## 4. Recommendation Function

The recommendation function identifies the selected movie, retrieves its cosine-similarity scores, sorts the movies from most similar to least similar, excludes the selected movie itself, and returns the requested number of recommendations.

```python
def recommend_movies(movie_title, number_of_recommendations=5):
    """
    Generate movie recommendations using cosine-similarity scores.

    Parameters:
        movie_title: The title selected by the user.
        number_of_recommendations: Number of recommendations to return.

    Returns:
        A pandas DataFrame containing ranked movie recommendations.
    """

    matching_movies = movies.index[
        movies["title"] == movie_title
    ].tolist()

    if not matching_movies:
        return pd.DataFrame({
            "Message": ["The selected movie was not found."]
        })

    selected_movie_index = matching_movies[0]

    similarity_scores = list(
        enumerate(
            cosine_similarity_matrix[selected_movie_index]
        )
    )

    similarity_scores = sorted(
        similarity_scores,
        key=lambda item: item[1],
        reverse=True
    )

    recommendations = []

    for movie_index, similarity_score in similarity_scores:
        if movie_index == selected_movie_index:
            continue

        movie = movies.iloc[movie_index]

        recommendations.append({
            "Rank": len(recommendations) + 1,
            "Movie": movie["title"],
            "Genre": movie["genre"],
            "Year": int(movie["year"]),
            "Rating": float(movie["rating"]),
            "Similarity (%)": round(
                float(similarity_score) * 100,
                2
            )
        })

        if len(recommendations) == number_of_recommendations:
            break

    return pd.DataFrame(recommendations)
```

## 5. Interactive User Interface

The interface allows users to choose a movie from a dropdown menu, specify the desired number of recommendations, and generate results by clicking a button. This provides a simple conversational-style interaction with the recommendation model.

```python
# Create the application title
application_title = widgets.HTML(
    value="""
    <h2>🎬 Smart Movie Recommendation System</h2>
    <p>
    Select a movie you enjoyed and the system will recommend
    similar movies using TF-IDF and cosine similarity.
    </p>
    """
)

# Create movie-selection dropdown
movie_dropdown = widgets.Dropdown(
    options=sorted(movies["title"].tolist()),
    value="Interstellar",
    description="Movie:",
    disabled=False,
    layout=widgets.Layout(width="600px"),
    style={"description_width": "100px"}
)

# Create recommendation-count slider
recommendation_slider = widgets.IntSlider(
    value=5,
    min=3,
    max=10,
    step=1,
    description="Results:",
    continuous_update=False,
    layout=widgets.Layout(width="600px"),
    style={"description_width": "100px"}
)

# Create recommendation button
recommend_button = widgets.Button(
    description="Generate Recommendations",
    button_style="success",
    tooltip="Click to generate movie recommendations",
    icon="search",
    layout=widgets.Layout(width="300px")
)

# Create output area
recommendation_output = widgets.Output()


def display_recommendations(button):
    """Display recommendations selected through the interface."""

    with recommendation_output:
        clear_output()

        selected_movie = movie_dropdown.value
        recommendation_count = recommendation_slider.value

        selected_details = movies[
            movies["title"] == selected_movie
        ].iloc[0]

        print("=" * 70)
        print(f"SELECTED MOVIE: {selected_movie}")
        print(f"GENRE: {selected_details['genre']}")
        print(f"YEAR: {int(selected_details['year'])}")
        print(f"RATING: {float(selected_details['rating']):.1f}/10")
        print("=" * 70)

        results = recommend_movies(
            selected_movie,
            recommendation_count
        )

        print(
            f"\nTop {recommendation_count} recommendations "
            f"for {selected_movie}:\n"
        )

        display(results)


recommend_button.on_click(display_recommendations)

display(
    application_title,
    movie_dropdown,
    recommendation_slider,
    recommend_button,
    recommendation_output
)
```
